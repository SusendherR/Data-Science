# 8)Test the similarity between the degree_t(Sci&Tech) and specialization(Mkt&HR) with respect to salary at significant level of 5%(Make decision using Hypothesis testing)


Independent Sample-Unpaired T Test. Different group(degree_t,specialisation)but same condition(salary)

```python
from scipy.stats import ttest_ind
#drop null values
dataset=dataset.dropna()
degree=dataset[dataset["degree_t"]=="Sci&Tech"]["salary"]
spec=dataset[dataset["specialisation"]=="Mkt&HR"]["salary"]

ttest_ind(degree,spec)
```

Hypothesis testing. Statement: Test the similarity between the degree_t (Sci & Tech) and specialization (Mkt & HR) with respect to salary. Alternative hypothesis: There is similarity between the degree_t of Sci & Tech and the specialization of Mkt & HR.

The p-value is less than 5%, so we accept the alternative hypothesis and reject the null hypothesis.

# 9)Convert the normal distribution to standardnormal distribution for salary columns

```python
def stdNBgraph(dataset):
    import seaborn as sns
    #calculate the mean,std
    mean=dataset.mean()
    std=dataset.std()
    # advantage of single line output converted to the list 
    values=[i for i in dataset]
    z_score=[((j-mean)/std) for j in values]
    # kde(Kernel density function)=True the curve will diply
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
```

# 10)What is the probability Density Function of the salary range from 700000 to 900000 ?

```python
def get_pdf_probability(dataset,startrange,endrange):
    # import the importance of library
    from matplotlib import pyplot
    # inbuilt function for normal distribution library
    from scipy.stats import norm
    # vizualization representation
    import seaborn as sns
    # distplot is kde(Kernel density function)=True the curve will diply ,false the curve will the be not disply
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'Green'},color='yellow')
    # axvline is virtical line startrange color
    pyplot.axvline(startrange,color='Red')
    # axvline is virtical line endrange color
    pyplot.axvline(endrange,color='Red')
    # above this line the graph creating 
    # Below lines for calculation
    # generate a sample
    sample=dataset
    # calculate the mean,std parameters 
    sample_mean=sample.mean()
    sample_std=sample.std()
    # print the mean ,std
    print('Mean=%.3f,Standard Deviation=%.3f'%(sample_mean,sample_std))
    # define the distribution
    # There is the normal distribution funcation.lear distributionPattern 
    dist=norm(sample_mean,sample_std)
    
    # sample probabilities for a range of outcomes
    # advantage of below line is in single line the output it  is converted into list it is call to one-liner for loop
    values = [value for value in range(startrange, endrange)]
    # whole probabilities summation the below lines
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["salary"],700000,900000)
```

# 11)Test the similarity between the degree_t(sci&Tech) with respect to etest_p and mba_p at significance level of 5%(Make decision using Hypothesis Testing)

Dependent sample-Paired T Test. Same group(degree_t)but Different condition(etest
_p,mba_p)

```python
from scipy.stats import ttest_ind
# null value drop
degree=dataset.dropna()
degree=dataset[dataset['degree_t']=='Sci&Tech']['etest_p']
degree1=dataset[dataset['degree_t']=='Sci&Tech']['mba_p']
# print(male)
ttest_ind(degree,degree1)
```

Hypothesis Report

H0 = There is no similarity between degree_t(Sci&Tech) with respect to etest_p and mba_p.

H1 = There is a similarity between degree_t(Sci&Tech) with respect to etest_p and mba_p.

Accept the null hypothesis and reject the alternative hypothesis.

The statistic = 4.53; there is a significant similarity between degree_t(Sci&Tech), etest_p, and mba_p, and the p-value < 0.05%.
